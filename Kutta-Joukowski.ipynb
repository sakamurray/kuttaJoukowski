# 講義「航空機概論」（空気力学の基礎）

圧力$p_\infty$, 速さ$V$の一様流中に置かれた半径$a$の円柱に働く力を計算する．

$$X = a \int_{0}^{2\pi} p(\theta) \cos \theta \mathrm{d} \theta$$

$$Y = -a \int_{0}^{2\pi} p(\theta) \sin \theta \mathrm{d} \theta$$

ここで$p$は円柱表面の局所圧力で，次式で与えられる．
$$p(\theta) = p_\infty + \frac{1}{2}\rho V^2 \left\{ 1 - \left( \frac{v(\theta)}{V} \right)^2 \right\}$$


```python
from IPython.display import Image
filename = "demoSymPy.png"
Image(filename)
```

```python
from sympy import *
init_printing()
```

```python
X, Y, theta, p, pf, rho, V, a, v, Gamma = symbols('X, Y, theta, p, pf, rho, V, a, v, Gamma')
```

*   循環がない場合
$$v(\theta) = 2V\sin \theta$$




```python
v = 2*V*sin(theta)
v
```

圧力分布
$$p(\theta) = p_\infty + \frac{1}{2}\rho V^2 \left\{ 1 - \left( \frac{v(\theta)}{V} \right)^2 \right\}$$

```python
p = pf + S(1)/2 * rho * V**2 * (1 - (v/V)**2)
p
```

圧力係数の分布

$$C_p = \frac{p-p_\infty}{\frac{1}{2}\rho V^2}$$

```python
Cp = (p-pf)/(S(1)/2 * rho * V**2)
Cp
plot(Cp, (theta, 0, 2*pi), ylabel="Cp")
```

$$X = a \int_{0}^{2\pi} p(\theta) \cos \theta \mathrm{d} \theta$$

```python
X = a * integrate(p*cos(theta), (theta,0,2*pi))
X
```

$$Y = -a \int_{0}^{2\pi} p(\theta) \sin \theta \mathrm{d} \theta$$

```python
Y = -a * integrate(p*sin(theta), (theta, 0, 2*pi))
Y
```

*   循環${\Gamma}$がある場合
$$v(\theta) = 2 V \sin \theta + \frac{\Gamma}{2 \pi a}$$



```python
v = 2 * V* sin(theta) + Gamma/(2*pi * a)
v
```

圧力分布
$$p(\theta) = p_\infty + \frac{1}{2}\rho V^2 \left\{ 1 - \left( \frac{v(\theta)}{V} \right)^2 \right\}$$

```python
p = pf + S(1)/2 * rho * V**2 * (1 - (v/V)**2)
p
```

圧力係数の分布

```python
Cp = (p-pf)/(S(1)/2 * rho * V**2)
Cp
expr= Cp.subs([(a, 1), (Gamma, 1), (V, 1)])
plot(expr, (theta, 0, 2*pi), ylabel="Cp")
```

$$X = a \int_{0}^{2\pi} p(\theta) \cos \theta \mathrm{d} \theta$$

```python
X = a * integrate(p*cos(theta), (theta,0,2*pi))
X
```

$$Y = -a \int_{0}^{2\pi} p(\theta) \sin \theta \mathrm{d} \theta$$

```python
Y = -a * integrate(p*sin(theta), (theta, 0, 2*pi))
Y
```
